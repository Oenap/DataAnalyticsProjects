```python
#Import Libraries (YolandaMokoena)

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv('/Users/yolanda/Downloads/movies.csv')
```

```python
 # Lets look at the data

df.head()
```

```python
# Searching for any missing data

for col in df.columns :
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,percent_missing))
```

```python
#Data types for the coloumns

df.dtypes
```

```python
#Changing the data type of the columns
df['budget'] = df['budget'].round().astype('Int64')
df['gross'] = df['gross'].round().astype('Int64')
```

```python
df.head()
```

```python
#correcting the year coloumn
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Drop any duplicated

df['company'].drop_duplicates().sort_values(ascending = False)
df['company'].head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes

df_numerized
df_numerized.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel( 'Movie Freatures')

plt.show() 
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs= corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
